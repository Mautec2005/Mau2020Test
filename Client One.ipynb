```python
import seaborn as sns
```

```python
fig = sns.FacetGrid(credit, hue='SEX',aspect=3)
fig.map(sns.kdeplot, 'LIMIT_BAL', shade=True)
oldest = credit['LIMIT_BAL'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Limit_Bal Grouped by Gender')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit, hue='EDUCATION',aspect=3)
fig.map(sns.kdeplot, 'LIMIT_BAL', shade=True)
oldest = credit['LIMIT_BAL'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Limit_Bal Grouped by Education')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit, hue='MARRIAGE',aspect=3)
fig.map(sns.kdeplot, 'LIMIT_BAL', shade=True)
oldest = credit['LIMIT_BAL'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Limit_Bal Grouped by Marriage')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit, hue='MARRIAGE',aspect=3)
fig.map(sns.kdeplot, 'BILL_AMT1', shade=True)
oldest = credit['BILL_AMT1'].max()
fig.set(title='Distribution of Bill_Status Grouped by Marriage')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit, hue='EDUCATION',aspect=3)
fig.map(sns.kdeplot, 'BILL_AMT1', shade=True)
oldest = credit['BILL_AMT1'].max()
fig.set(title='Distribution of Bill_Status Grouped by Education')
fig.add_legend()
```

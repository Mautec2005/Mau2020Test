```python
#features
features = credit.iloc[:,0:24]
print('Summary of feature sample')
features.head()
```

```python
#dependent variable
depVar = credit['Default_Y/N']
```

```python
X_train = (features[: 30000])
y_train = depVar[: 30000]
X_test = features[-3000:]
y_test = depVar[-3000:]
```

```python
#cross validation
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train)
```

```python
X_train.shape, X_test.shape
```

```python
import numpy as np
import scipy
from math import sqrt
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Models
modelknn = KNeighborsClassifier(n_neighbors=20, weights='uniform')
modelknn.fit(X_train, y_train)
```

```python
print(cross_val_score(modelknn, X_train, y_train))
```

```python
modelknn.score(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modeldtree = DecisionTreeClassifier(max_depth=50, random_state=101,
                                   max_features = None, min_samples_leaf = 15)
modeldtree.fit(X_train, y_train)
```

```python
print(cross_val_score(modeldtree, X_train, y_train))
```

```python
modeldtree.score(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelRF = RandomForestClassifier(n_estimators = 70, oob_score = True, n_jobs = -1,
                                random_state = 101, max_features = None, min_samples_leaf = 30)
modelRF.fit(X_train, y_train)
```

```python
print(cross_val_score(modelRF, X_train, y_train))
```

```python
modelRF.score(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelLR = LogisticRegression()
modelLR.fit(X_train, y_train)
```

```python
print(cross_val_score(modelLR, X_train, y_train))
```

```python
modelLR.score(X_train,y_train)
```

```python
# Cross Validation Classification Report
#import pandas
from sklearn import model_selection
from sklearn.metrics import classification_report
```

```python
#Make Predictions
predictions = modelknn.predict(X_test)
report = classification_report(y_test, predictions)
print(report)
```

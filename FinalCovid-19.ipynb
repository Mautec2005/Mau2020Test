```python
confirmedCases = pd.read_csv (r'C:\Users\malvarez4\Desktop\All\Personal Things\MICITT\Curso\5.Data Science with Python\Task 4 Complete a Data Science Capstone Project\time_series_covid_19_confirmed.csv')
```

```python
confirmedCases.head()
```

```python
columns = confirmedCases.keys()
```

```python
columns
```

```python
confirmedDays = confirmedCases.loc[:, columns[4]:columns[-1]]
```

```python
confirmedDays.head()
```

```python
dates = confirmedDays.keys()
```

```python
casesAroundWorld = []
for i in dates:
    sumOfConfirmedCases = confirmedDays[i].sum()
    casesAroundWorld.append(sumOfConfirmedCases)
```

```python
sumOfConfirmedCases
```

```python
numberOfDays = np.array([i for i in range(len(dates))]).reshape(-1, 1)
```

```python
numberOfDays
```

```python
casesAroundWorld
```

```python
numberOfDays = np.array([i for i in range(len(dates))]).reshape(-1, 1)
casesAroundWorld = np.array(casesAroundWorld).reshape(-1, 1)
```

```python
#15 days forecast
numberFutureDays = 15
forecast = np.array([i for i in range(len(dates)+numberFutureDays)]).reshape(-1, 1)
```

```python
forecast
```

```python
import datetime
```

```python
#start = '1/22/20'
startDate = datetime.datetime.strptime('1/22/20', '%m/%d/%y')
forecastDates = []
for i in range(len(forecast)):
    forecastDates.append((startDate + datetime.timedelta(days=i)).strftime('%m/%d/%y'))
```

```python
confirmsLastDay = confirmedCases[dates[-1]]
```

```python
confirmsLastDay
```

```python
countriesList = list(confirmedCases['Country/Region'].unique())
```

```python
import operator
```

```python
confirmedCasesPerCountry = []
withoutCases = []
for i in countriesList:
    cases = confirmsLastDay[confirmedCases['Country/Region']==i].sum()
    if cases > 0:
        confirmedCasesPerCountry.append(cases)
    else:
        withoutCases.append(i)
for i in withoutCases:
    countriesList.remove(i)
countriesList = [k for k, v in sorted(zip(countriesList, confirmedCasesPerCountry),key=operator.itemgetter(1), reverse=True)]
for i in range(len(countriesList)):
    confirmedCasesPerCountry[i] = confirmsLastDay[confirmedCases['Country/Region']==countriesList[i]].sum()
```

```python
confirmedCasesPerCountry
```

```python
print('Confirmed Cases:')
for i in range(len(countriesList)):
    print(f'{countriesList[i]}: {confirmedCasesPerCountry[i]} cases')
```

```python
plt.figure(figsize=(10, 30))
plt.barh(countriesList, confirmedCasesPerCountry)
plt.title('Confirmed Cases per Countries')
plt.xlabel('Number of Confirmed Cases')
plt.show()
```

```python
top10Countries = []
top10ConfirmedCases = []
others = np.sum(confirmedCasesPerCountry[10:])
for i in range(len(confirmedCasesPerCountry[:10])):
    top10Countries.append(countriesList[i])
    top10ConfirmedCases.append(confirmedCasesPerCountry[i])
top10Countries.append('Others')
top10ConfirmedCases.append(others)
```

```python
top10Countries
```

```python
top10ConfirmedCases
```

```python
plt.figure(figsize=(20, 10))
plt.barh(top10Countries, top10ConfirmedCases)
plt.title('Top10 confirmed Cases', size=20)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(numberOfDays, casesAroundWorld, test_size=0.15, shuffle=False)
```

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#SVM Model
kernel = ['poly', 'sigmoid', 'rbf']
c = [0.01, 0.1, 1, 10]
gamma = [0.01, 0.1, 1]
epsilon = [0.01, 0.1, 1]
shrinking = [True, False]
svm_grid = {'kernel': kernel, 'C': c, 'gamma' : gamma, 'epsilon': epsilon, 'shrinking' : shrinking}

svm = SVR()
svmModel = RandomizedSearchCV(svm, svm_grid, scoring='neg_mean_squared_error', cv=3, return_train_score=True, n_jobs=-1, n_iter=40, verbose=1)
svmModel.fit(X_train, np.ravel(y_train, order='C'))
```

```python
svmModel.best_params_
```

```python
svmConfirmed = svmModel.best_estimator_
svmPrediction = svmConfirmed.predict(forecast)
```

```python
svmConfirmed
```

```python
svmPrediction
```

```python
testSvmPrediction = svmConfirmed.predict(X_test)
plt.plot(testSvmPrediction)
plt.plot(y_test)
print('Mean Absolute Error:', mean_absolute_error(testSvmPrediction, y_test))
print('Mean Squared Error:', mean_squared_error(testSvmPrediction, y_test))
```

```python
plt.figure(figsize=(20, 10))
plt.plot(numberOfDays, casesAroundWorld)
plt.title('Covid-19 cases Over time', size=30)
plt.xlabel('Days since January 22', size=30)
plt.ylabel('Number of cases', size=30)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(numberOfDays, casesAroundWorld)
plt.plot(forecast, svmPrediction, linestyle='dashed', color='red')
plt.title('Covid-19 cases Over time', size=30)
plt.xlabel('Days since January 22', size=30)
plt.ylabel('Number of cases', size=30)
plt.legend(['Confirmed Cases', 'SVM Predictions'])
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

```python
rfRegModel = RandomForestRegressor(n_estimators=200, random_state=50)
rfRegModel.fit(X_train, np.ravel(y_train, order='C'))
rfPrediction = rfRegModel.predict(X_test)
```
